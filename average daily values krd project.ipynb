```python
import openpyxl
import matplotlib.pyplot as plt

# Открываем файл Excel
file_path = "C:/Diploma.xlsx"  # Замените на фактический путь к вашему файлу
workbook = openpyxl.load_workbook(file_path)

# Получаем лист с данными
sheet = workbook.worksheets[1]  # Или используйте workbook.worksheets[0], если ваш лист не является активным

# Получаем данные из столбца с годовыми почасовыми данными (например, столбец M)
yearly_data = [float(cell.value) for cell in sheet['O'][1:8761] if cell.value is not None]

# Вычисляем средние значения для каждого часа в сутках
daily_averages = [sum(yearly_data[i::24]) / len(yearly_data[i::24]) for i in range(24)]

# Выводим почасовые средние данные в строку вывода
for hour, average in enumerate(daily_averages, start=1):
    print(f"{average}")

# Создаем график
plt.plot(range(1, 25), daily_averages)

# Подписываем график
plt.title("Среднесуточный график")
plt.xlabel("Час суток")
plt.ylabel("Среднее значение")

# Выводим график
plt.show()


```
